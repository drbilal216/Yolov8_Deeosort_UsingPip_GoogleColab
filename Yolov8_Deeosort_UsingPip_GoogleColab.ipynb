```python
import cv2
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
from time import time

model = YOLO("yolov8x.pt")

video_path = "3.mp4"
cap = cv2.VideoCapture(video_path)

# Get the video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

# Define the codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_path = 'output.mp4'
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

tracker = DeepSort(max_age=70)

while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        #results = model(frame, device='cpu', classes=0, conf=0.3)
        results = model(frame, device='0', classes=0, conf=0.3)
        #annotated_frame = results[0].plot()

        # tracker
        bbs = []

        if results[0].boxes.xywh.tolist():
            for d in results:
                 for box, conf, cls in zip(d.boxes.xywh.tolist(), d.boxes.conf.tolist(), d.boxes.cls.tolist()):
                    bb = (box, conf, cls)
                    bbs.append(bb)

        tracks = tracker.update_tracks(bbs, frame=results[0].orig_img)

        for track in tracks:
            track_id = track.track_id
            cx, cy, width, height = track.to_tlwh()
            x1 = int(cx - width / 2)
            y1 = int(cy - height / 2)
            x2 = int(cx + width / 2)
            y2 = int(cy + height / 2)

            text = f"ID: {track_id}"
            cv2.rectangle(results[0].orig_img, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(results[0].orig_img, text, ((x1), (y1) - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv2.LINE_AA)

        # Write the frame to the output video file
        out.write(cv2.cvtColor(results[0].orig_img, cv2.COLOR_RGB2BGR))

        #cv2.imshow("Video", results[0].orig_img)
        if cv2.waitKey(1) & 0xFF == ord("q"):
            break
    else:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
